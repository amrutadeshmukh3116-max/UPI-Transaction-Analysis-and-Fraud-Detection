## EDA

```python
# Trends over year:
df1=pd.read_csv('upi_transaction_history.csv')
df1
df1['date_only']=pd.to_datetime(df1['timestamp'],format='%d-%b-%y %H:%M:%S',errors='coerce')
df1['date_only']=df1['date_only'].dt.year
df1.head(10)

df_year=df1.groupby('year')['amount'].mean().reset_index()

sns.lineplot(data=df_year,x='year',y='amount')
plt.show()
```

Conclusion-From the above visual,we can say that the amount rises sharply in 2021 compared to 2020,indicating a surge in transaction value.The highest average amount is recorded in 2021,making it the peak year in the given time range.There is a noticeable decline from 2021 to 2022,which indicates operational changes.From 2022 to 2025,the amount shows only minor fluctuations.The trend line is relatively flat after 2022,which implies consistent transaction behaviour over recent years.

```python
# Trends over month:
df1=pd.read_csv('upi_transaction_history.csv')
df1['date_only']=pd.to_datetime(df1['timestamp'],format='%d-%b-%y %H:%M:%S',errors='coerce')
df1['date_only']=df1['date_only'].dt.month
df1.head(10)
df_month=df1.groupby('month')['amount'].mean().reset_index()
sns.lineplot(data=df_month,x='month',y='amount')
plt.show()
```

conclusion-The average amount varies slightly month to month,indicating stable but seasonally influenced spending behaviour.Highest average amount around month 3 and month 9 so,these are high-usage months.A noticeable dip occurs in month 10,indicating reduced transaction value.The overall range of values is narrow (around 42-43),which suggests consistent transaction patterns rather than erratic behaviour.Repeated rises and falls imply that transaction amount is influenced by monthly cycles rather than random variation.

```python
# Heatmap:
region_fraud=txn_region.pivot_table(
    values='fraud_flag',
    index='region',
    columns='transaction_type',
    aggfunc='mean'
)
device_fraud=df1.pivot_table(
    values='fraud_flag',
    index='device_type',
    columns='channel',
    aggfunc='mean'
)
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.heatmap(region_fraud,annot=True,fmt='.3f',cmap='Reds')
plt.title("fraud rate by region & transaction type")

plt.subplot(1,2,2)
sns.heatmap(device_fraud,annot=True,fmt='.3f',cmap='Reds')
plt.title("fraud by device & channel")
plt.show()
```

Conclusion- West region is the highest fraud-prone region.It shows the darkest colour overall.Especially high for Bill Pay indicates west region has systematically higher fraud risk.Bill Pay is the riskiest transaction type across regions.Bill pay transactions need extra verification rules.Merchant Payment also shows elevated fraud in central and east.Receive transactions are comparatively safer.South and North regions have moderate and stable fraud rates.

QR Code channel is the riskiest overall.Feature phones have elevated fraud risk across all channels,suggesting weaker security or higher misuse on these devices.Android has moderate to high fraud especially on app channel.ios is the safest device overall.Fraud risk increases when moving from intent-based flows to QR-based payments.

## Statistical tests:

ANOVA to evaluate variability in fraud rates by merchant type and region 

Null hypothesis(Ho): All groups have the same fraud rate

Alternate hypothesis(H1): All groups have the different fraud rate

```python
from scipy.stats import f_oneway 

txn_merchant=df1.merge(df7[['merchant_id','merchant_type']],
                      on='merchant_id',
                      how='left'
                      )
groups=[group['fraud_flag'].values
        for name,group in txn_merchant.groupby('merchant_type')
      ]
f_stat,p_value=f_oneway(*groups)

print("f-statistics:",f_stat)
print("P-value:",p_value)
```

ANOVA Test to compare average transaction amounts across device types

Null hypothesis(Ho): Average transaction amount is the same across device types

Alternate hypothesis(H1): Average transaction amount is different

```python
groups=[group['amount'].values
        for name,group in txn_merchant.groupby('device_type')
      ]
f_stat,p_value=f_oneway(*groups)

print("f-statistics:",f_stat)
print("P-value:",p_value)
```

```python
# chi-square test:
import pandas as pd
from scipy.stats import chi2_contingency
df=pd.read_csv('upi_transaction_history.csv')
tabel1=pd.crosstab(df1['fraud_flag'],df1['channel'])
chi2,p,dof,expected=chi2_contingency(tabel1)
print("Fraud flag vs channel")
print("chi-square:",chi2)
print("p-value:",p)
table2=pd.crosstab(df1['status'],df1['device_type'])
chi2,p,dof,expected=chi2_contingency(table2)
print("\nstatus vs Device type")
print("chi-square:",chi2)
print("p-value:",p)
```

Conclusion-p-value(0.143)>0.05,there is no statistically significant relationship between fraud flag and transaction channel.Fraud occurence does not depend strongly on the channel.
p-value(0.431)>0.05, there is no statistically significant relationship between transaction status and device type.Transaction success/failure/pending is not influenced by the type of device.

```python
# Correlation analysis:
import pandas as pd
risk_score=pd.read_csv('customer_master.csv')
fraud_flag=pd.read_csv('upi_transaction_history.csv')
df=pd.merge(risk_score,fraud_flag,on='customer_id',how='inner')
print(df.head())
df_corr=df[['risk_score','fraud_flag']].copy()
df_corr.dropna(inplace=True)
df_corr['fraud_flag']=df_corr['fraud_flag'].map({False:0,True:1})
corr_matrix=df_corr.corr()

plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',fmt='.2f',square=True)
plt.title("corelation between risk score and fraud flag")
plt.show()

from scipy.stats import pointbiserialr
corr,p_value=pointbiserialr(df_corr['fraud_flag'],df_corr['risk_score'])
print("coorelation coefficient:",corr)
print("P-value:",p_value)
```

Insight: The corelation coefficient between risk_score and fraud_flag is -0.0016,which is almost zero.
The p-value=0.612,which is much higher than 0.05
This means,There is no statistically significant relationship between risk_score and fraud_flag is -0.0016,which is almost zero.
